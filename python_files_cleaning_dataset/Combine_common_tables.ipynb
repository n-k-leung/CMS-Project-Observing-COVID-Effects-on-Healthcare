combining hcpcs_codes tables

```python
import pandas as pd

# Load the two CSV files
df1 = pd.read_csv('hcpcs_codes.csv')
df2 = pd.read_csv('hcpcs_codes1.csv')

assert df1.columns.equals(df2.columns), "The two tables have different columns"
```

```python
# Combine the two DataFrames
combined_df = pd.concat([df1, df2], ignore_index=True)

# Drop exact duplicates (all columns)
deduped_df = combined_df.drop_duplicates()

# (Optional) Save the cleaned result
deduped_df.to_csv('hcpcs_combined.csv', index=False)
print("combined!")
```

combining provider_address

```python
import pandas as pd

# Define dtype spec to avoid mixed type warning
dtype_spec = {'Rndrng_Prvdr_State_FIPS': str}

# Load both CSV files with correct dtype
df1 = pd.read_csv('provider_address.csv', dtype=dtype_spec)
df2 = pd.read_csv('provider_address1.csv', dtype=dtype_spec)

# Reorder df2 columns to match df1
df2 = df2[df1.columns]

# Check that columns match (ignoring order)
assert set(df1.columns) == set(df2.columns), "The two tables have different columns"
```

```python
# Combine the two DataFrames
combined_df = pd.concat([df1, df2], ignore_index=True)

# Drop exact duplicates (all columns)
deduped_df = combined_df.drop_duplicates()

# (Optional) Save the cleaned result
deduped_df.to_csv('provider_address_combined.csv', index=False)
print("combined!")
```
