```python
#we are using pandas for data manipulation. It is a powerful library for handling structured data. Otherwise,
#  we would have to write a lot of code! .
import pandas as pd 
```

```python
#Here we are creating a dataframe for the year 2023
#We can consider the dataframe (df_2023) as a table where all the data from Geography_and_Service_2023 has been loaded. 
file_name_Providers_Service_2023 = 'MUP_PHY_R25_P05_V20_D23_Prov_Svc.csv'
df_2023 = pd.read_csv(file_name_Providers_Service_2023, low_memory=False)

# Here we are adding column (year) to our 2023's dataset
df_2023['Year'] = 2023

#Here seeing the column names and first 3 rows.
df_2023.head(3) 
```

```python
#Here we are creating a dataframe for the year 2019
#We can consider the dataframe (df_2019) as a table where all the data from Geography_and_Service_2019 has been loaded. 
file_name_Providers_Service_2019 = 'MUP_PHY_R25_P04_V10_D19_Prov_Svc.csv'
df_2019 = pd.read_csv(file_name_Providers_Service_2019, low_memory=False)

# Here we are adding column (year) to our 2019's dataset
df_2019['Year'] = 2019

#Here seeing the column names and first 3 rows.
df_2019.head(3) 
```

```python
#Here we are doing concatenation which means, combining two dataframes (df_2023, df_2019) and creating a new dataframe (df),
#which we can consider as a new table 
df = pd.concat([df_2023, df_2019], axis=0)

# df.head() this function only shows the first 5 rows
df.head()
```

```python
#Here by using this format we are defining the datatypes to each of our columns
#If we want to add a string datatype, make sure using 'String'. for float and INT, it will be fine as it is.
df['Rndrng_Prvdr_Last_Org_Name'] = df['Rndrng_Prvdr_Last_Org_Name'].astype('string')
df['Rndrng_Prvdr_First_Name'] = df['Rndrng_Prvdr_First_Name'].astype('string')
df['Rndrng_Prvdr_MI'] = df['Rndrng_Prvdr_MI'].astype('string')
df['Rndrng_Prvdr_Crdntls'] = df['Rndrng_Prvdr_Crdntls'].astype('string')
df['Rndrng_Prvdr_Ent_Cd'] = df['Rndrng_Prvdr_Ent_Cd'].astype('string')
df['Rndrng_Prvdr_St1'] = df['Rndrng_Prvdr_St1'].astype('string')
df['Rndrng_Prvdr_St2'] = df['Rndrng_Prvdr_St2'].astype('string')
df['Rndrng_Prvdr_City'] = df['Rndrng_Prvdr_City'].astype('string')

df['Rndrng_Prvdr_State_Abrvtn'] = df['Rndrng_Prvdr_State_Abrvtn'].astype('string')
df['Rndrng_Prvdr_State_FIPS'] = df['Rndrng_Prvdr_State_FIPS'].astype('string')
# df['Rndrng_Prvdr_Zip5'] = df['Rndrng_Prvdr_Zip5'].astype('int') has K1H8' (which looks like a Canadian postal code)
df['Rndrng_Prvdr_Zip5'] = pd.to_numeric(df['Rndrng_Prvdr_Zip5'], errors='coerce')
df['Rndrng_Prvdr_RUCA_Desc'] = df['Rndrng_Prvdr_RUCA_Desc'].astype('string')
df['Rndrng_Prvdr_Cntry'] = df['Rndrng_Prvdr_Cntry'].astype('string')
df['Rndrng_Prvdr_Type'] = df['Rndrng_Prvdr_Type'].astype('string')
df['Rndrng_Prvdr_Mdcr_Prtcptg_Ind'] = df['Rndrng_Prvdr_Mdcr_Prtcptg_Ind'].astype('string')

df['HCPCS_Cd'] = df['HCPCS_Cd'].astype('string')
df['HCPCS_Desc'] = df['HCPCS_Desc'].astype('string')
df['HCPCS_Drug_Ind'] = df['HCPCS_Drug_Ind'].astype('string')
df['Place_Of_Srvc'] = df['Place_Of_Srvc'].astype('string')

pd.set_option('display.max_columns', None)  # Show all columns
# print(df.dtypes)
print(df.dtypes.to_string())
```

```python
#If a cell contains a missing value (like NaN or None), it returns True.
#If a cell contains a valid value, it returns False.
#To understand which column has null value or not we can write this funtion and print. 
#(We might need to understand this when we will decide to write query/ might need to handle null values).

import numpy as np

# replace empty strings with NaN
df.replace('', np.nan, inplace=True)

# check columns with any nulls
null_columns = df.columns[df.isnull().any()]
print("Columns with null values:", null_columns.tolist())

# count of nulls per column
print(df.isnull().sum()[df.isnull().sum() > 0])
```

```python
# Provider basic info
provider_cols = [
    'Rndrng_NPI', 'Rndrng_Prvdr_Last_Org_Name', 'Rndrng_Prvdr_First_Name', 'Rndrng_Prvdr_MI',
    'Rndrng_Prvdr_Crdntls', 'Rndrng_Prvdr_Ent_Cd', 'Year'
]

# Provider Address
address_cols = [
    'Rndrng_NPI', 'Rndrng_Prvdr_St1', 'Rndrng_Prvdr_St2', 'Rndrng_Prvdr_City',
    'Rndrng_Prvdr_State_Abrvtn', 'Rndrng_Prvdr_Zip5', 'Rndrng_Prvdr_Cntry'
]
# Ruca
ruca_cols = [
    'Rndrng_Prvdr_RUCA', 'Rndrng_Prvdr_RUCA_Desc'
]

# State
state_cols = [
    'Rndrng_Prvdr_State_Abrvtn', 'Rndrng_Prvdr_State_FIPS'
]
# Provider Classification
classification_cols = [
    'Rndrng_NPI', 'Rndrng_Prvdr_RUCA','Rndrng_Prvdr_Type', 'Rndrng_Prvdr_Mdcr_Prtcptg_Ind', 'Year'
]

# HCPCS Table
hcpcs_cols = [
    'HCPCS_Cd', 'HCPCS_Desc', 'HCPCS_Drug_Ind', 'Place_Of_Srvc'
]

# Benefit Service and Amount Costs per provider
benefit_service_amount_cols = [
    'Rndrng_NPI', 'HCPCS_Cd', 'Year', 'Tot_Benes', 'Tot_Srvcs',
    'Tot_Bene_Day_Srvcs', 'Avg_Sbmtd_Chrg', 'Avg_Mdcr_Alowd_Amt',
    'Avg_Mdcr_Pymt_Amt', 'Avg_Mdcr_Stdzd_Amt'
]

# drop duplicated Rndrng_NPI, which is our primary key, only if year is the same, because we have data from year 2019 and 2023
df_provider = df[provider_cols].drop_duplicates(subset=['Rndrng_NPI', 'Year']).reset_index(drop=True)
df_address = df[address_cols].drop_duplicates(subset=['Rndrng_NPI']).reset_index(drop=True)
df_ruca = df[ruca_cols].drop_duplicates(subset=['Rndrng_Prvdr_RUCA']).reset_index(drop=True)
df_state = df[state_cols].drop_duplicates(subset=['Rndrng_Prvdr_State_Abrvtn']).reset_index(drop=True)
df_classification = df[classification_cols].drop_duplicates(subset=['Rndrng_NPI', 'Year']).reset_index(drop=True)
df_hcpcs = df[hcpcs_cols].drop_duplicates(subset=['HCPCS_Cd']).reset_index(drop=True)
df_benefit_service_amount = df[benefit_service_amount_cols].drop_duplicates(subset=['Rndrng_NPI', 'Year', 'HCPCS_Cd']).reset_index(drop=True)
```

```python
# Export each DataFrame to CSV
df_provider.to_csv("provider1.csv", index=False)
df_address.to_csv("provider_address1.csv", index=False)
df_ruca.to_csv('ruca1.csv', index=False)
df_state.to_csv('state1.csv', index=False)
df_classification.to_csv("provider_classification1.csv", index=False)
df_hcpcs.to_csv("hcpcs_codes.csv", index=False)
df_benefit_service_amount.to_csv("provider_benefit_service_amounts.csv", index=False)

print("All normalized tables exported successfully!")
```
