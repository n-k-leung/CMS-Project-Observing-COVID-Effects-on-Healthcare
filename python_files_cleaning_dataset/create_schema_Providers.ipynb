```python
#we are using pandas for data manipulation. It is a powerful library for handling structured data. Otherwise,
#  we would have to write a lot of code! .
# !pip3 install pandas
import pandas as pd 
```

```python
#Here we are creating a dataframe for the year 2023
#We can consider the dataframe (df_2023) as a table where all the data from Geography_and_Service_2023 has been loaded. 
file_name_Providers_2023 = 'MUP_PHY_R25_P05_V20_D23_Prov.csv'
df_2023 = pd.read_csv(file_name_Providers_2023, low_memory=False)

# Here we are adding column (year) to our 2023's dataset
df_2023['Year'] = 2023

#Here seeing the column names and first 3 rows.
df_2023.head(3) 
```

```python
#Here we are creating a dataframe for the year 2019
#We can consider the dataframe (df_2019) as a table where all the data from Geography_and_Service_2019 has been loaded. 
file_name_Providers_2019 = 'MUP_PHY_R25_P07_V10_D19_Prov.csv'
df_2019 = pd.read_csv(file_name_Providers_2019, low_memory=False)

# Here we are adding column (year) to our 2019's dataset
df_2019['Year'] = 2019

#Here seeing the column names and first 3 rows.
df_2019.head(3) 
```

```python
#Here we are doing concatenation which means, combining two dataframes (df_2023, df_2019) and creating a new dataframe (df),
#which we can consider as a new table 
df = pd.concat([df_2023, df_2019], axis=0)

# df.head() this function only shows the first 5 rows
df.head()
```

```python
#Here we will be doing the data cleaning part. The columns below have $ and , which we need to remove. 
#Otherwise it will be considered as a string. We need this step to define the data type for our SQL table.

# cols_to_clean = ['Tot_Rndrng_Prvdrs', 'Tot_Benes', 'Tot_Srvcs', 'Tot_Bene_Day_Srvcs', 'Avg_Sbmtd_Chrg',
#                  'Avg_Mdcr_Alowd_Amt', 'Avg_Mdcr_Pymt_Amt', 'Avg_Mdcr_Stdzd_Amt' ]

# for col in cols_to_clean:
#     df[col] = df[col].str.replace('$', '', regex=False)
#     df[col] = df[col].str.replace(',', '', regex=False)

#here we are printing the top five rows just to check if the $ and commas have been removed or or not. 
#And after running the cell we will see it does
# df[0:5]
```

```python
#Before defining the data type, 
#if we run this cell we will see, the datatypes for the columns are not showing correctly
#But after running cell [197], we will run this cell again and make sure all the datatypes are correctly added or not

pd.set_option('display.max_columns', None)  # Show all columns
# print(df.dtypes)
print(df.dtypes.to_string())
```

```python
#Here by using this format we are defining the datatypes to each of our columns
#If we want to add a string datatype, make sure using 'String'. for float and INT, it will be fine as it is.
df['Rndrng_Prvdr_Last_Org_Name'] = df['Rndrng_Prvdr_Last_Org_Name'].astype('string')
df['Rndrng_Prvdr_First_Name'] = df['Rndrng_Prvdr_First_Name'].astype('string')
df['Rndrng_Prvdr_MI'] = df['Rndrng_Prvdr_MI'].astype('string')
df['Rndrng_Prvdr_Crdntls'] = df['Rndrng_Prvdr_Crdntls'].astype('string')
df['Rndrng_Prvdr_Ent_Cd'] = df['Rndrng_Prvdr_Ent_Cd'].astype('string')
df['Rndrng_Prvdr_St1'] = df['Rndrng_Prvdr_St1'].astype('string')
df['Rndrng_Prvdr_St2'] = df['Rndrng_Prvdr_St2'].astype('string')
df['Rndrng_Prvdr_City'] = df['Rndrng_Prvdr_City'].astype('string')

df['Rndrng_Prvdr_State_Abrvtn'] = df['Rndrng_Prvdr_State_Abrvtn'].astype('string')
df['Rndrng_Prvdr_State_FIPS'] = df['Rndrng_Prvdr_State_FIPS'].astype('string')
# df['Rndrng_Prvdr_Zip5'] = df['Rndrng_Prvdr_Zip5'].astype('int') has K1H8' (which looks like a Canadian postal code)
df['Rndrng_Prvdr_Zip5'] = pd.to_numeric(df['Rndrng_Prvdr_Zip5'], errors='coerce')
df['Rndrng_Prvdr_RUCA_Desc'] = df['Rndrng_Prvdr_RUCA_Desc'].astype('string')
df['Rndrng_Prvdr_Cntry'] = df['Rndrng_Prvdr_Cntry'].astype('string')
df['Rndrng_Prvdr_Type'] = df['Rndrng_Prvdr_Type'].astype('string')
df['Rndrng_Prvdr_Mdcr_Prtcptg_Ind'] = df['Rndrng_Prvdr_Mdcr_Prtcptg_Ind'].astype('string')

df['Drug_Sprsn_Ind'] = df['Drug_Sprsn_Ind'].astype('string')
df['Med_Sprsn_Ind'] = df['Med_Sprsn_Ind'].astype('string')

pd.set_option('display.max_columns', None)  # Show all columns
# print(df.dtypes)
print(df.dtypes.to_string())
```

```python
df['Med_Sprsn_Ind'] = df['Med_Sprsn_Ind'].replace({
    '*': 'Suppressed: fewer than 11 beneficiaries',
    '#': 'Counter-suppressed to prevent re-calculation',
    '': 'Not suppressed'
}).fillna('Not suppressed')

df['Drug_Sprsn_Ind'] = df['Drug_Sprsn_Ind'].replace({
    '*': 'Suppressed: fewer than 11 beneficiaries',
    '#': 'Counter-suppressed to prevent re-calculation',
    '': 'Not suppressed'
}).fillna('Not suppressed')
```

```python
#The format of the tuple is always (number of rows, number of columns).
# it is Our original dataframe(or consider it as data table which has 537268 rows and 16 columns )
df.shape
```

```python
#If a cell contains a missing value (like NaN or None), it returns True.
#If a cell contains a valid value, it returns False.
#To understand which column has null value or not we can write this funtion and print. 
#(We might need to understand this when we will decide to write query/ might need to handle null values).

import numpy as np

# replace empty strings with NaN
df.replace('', np.nan, inplace=True)

# check columns with any nulls
null_columns = df.columns[df.isnull().any()]
print("Columns with null values:", null_columns.tolist())

# count of nulls per column
print(df.isnull().sum()[df.isnull().sum() > 0])
```

```python
# Provider basic info
provider_cols = [
    'Rndrng_NPI', 'Rndrng_Prvdr_Last_Org_Name', 'Rndrng_Prvdr_First_Name', 'Rndrng_Prvdr_MI',
    'Rndrng_Prvdr_Crdntls', 'Rndrng_Prvdr_Ent_Cd', 'Year'
]

# Provider Address
address_cols = [
    'Rndrng_NPI', 'Rndrng_Prvdr_St1', 'Rndrng_Prvdr_St2', 'Rndrng_Prvdr_City', 'Rndrng_Prvdr_State_Abrvtn',
    'Rndrng_Prvdr_Zip5', 'Rndrng_Prvdr_Cntry'
]

# Ruca
ruca_cols = [
    'Rndrng_Prvdr_RUCA', 'Rndrng_Prvdr_RUCA_Desc'
]

# State
state_cols = [
    'Rndrng_Prvdr_State_Abrvtn', 'Rndrng_Prvdr_State_FIPS'
]

# Overall Service Totals
service_totals_cols = [
    'Rndrng_NPI', 'Tot_HCPCS_Cds', 'Tot_Benes', 'Tot_Srvcs', 'Tot_Sbmtd_Chrg', 'Tot_Mdcr_Alowd_Amt', 'Tot_Mdcr_Pymt_Amt', 
    'Tot_Mdcr_Stdzd_Amt', 'Year'

]

# Provider Classification
classification_cols = [
    'Rndrng_NPI', 'Rndrng_Prvdr_RUCA','Rndrng_Prvdr_Type', 'Rndrng_Prvdr_Mdcr_Prtcptg_Ind', 'Year'
]

# Drug Services
drug_services_cols = [
    'Rndrng_NPI', 'Drug_Sprsn_Ind', 'Drug_Tot_HCPCS_Cds', 'Drug_Tot_Benes', 'Drug_Tot_Srvcs',
    'Drug_Sbmtd_Chrg', 'Drug_Mdcr_Alowd_Amt', 'Drug_Mdcr_Pymt_Amt', 'Drug_Mdcr_Stdzd_Amt', 'Year'
]

# Medical Services
medical_services_cols = [
    'Rndrng_NPI', 'Med_Sprsn_Ind', 'Med_Tot_HCPCS_Cds', 'Med_Tot_Benes', 'Med_Tot_Srvcs',
    'Med_Sbmtd_Chrg', 'Med_Mdcr_Alowd_Amt', 'Med_Mdcr_Pymt_Amt', 'Med_Mdcr_Stdzd_Amt', 'Year'
]

# Beneficiary Demographics
bene_demo_cols = [
    'Rndrng_NPI', 'Bene_Avg_Age', 'Bene_Age_LT_65_Cnt', 'Bene_Age_65_74_Cnt',
    'Bene_Age_75_84_Cnt', 'Bene_Age_GT_84_Cnt', 'Bene_Feml_Cnt', 'Bene_Male_Cnt',
    'Bene_Race_Wht_Cnt', 'Bene_Race_Black_Cnt', 'Bene_Race_API_Cnt', 'Bene_Race_Hspnc_Cnt',
    'Bene_Race_NatInd_Cnt', 'Bene_Race_Othr_Cnt', 'Bene_Dual_Cnt', 'Bene_Ndual_Cnt', 'Year'
]

# Beneficiary Clinical Conditions
bene_cc_cols = [
    'Rndrng_NPI', 'Bene_CC_BH_ADHD_OthCD_V1_Pct', 'Bene_CC_BH_Alcohol_Drug_V1_Pct', 'Bene_CC_BH_Tobacco_V1_Pct',
    'Bene_CC_BH_Alz_NonAlzdem_V2_Pct', 'Bene_CC_BH_Anxiety_V1_Pct', 'Bene_CC_BH_Bipolar_V1_Pct',
    'Bene_CC_BH_Mood_V2_Pct', 'Bene_CC_BH_Depress_V1_Pct', 'Bene_CC_BH_PD_V1_Pct', 'Bene_CC_BH_PTSD_V1_Pct',
    'Bene_CC_BH_Schizo_OthPsy_V1_Pct', 'Bene_CC_PH_Asthma_V2_Pct', 'Bene_CC_PH_Afib_V2_Pct',
    'Bene_CC_PH_Cancer6_V2_Pct', 'Bene_CC_PH_CKD_V2_Pct', 'Bene_CC_PH_COPD_V2_Pct', 'Bene_CC_PH_Diabetes_V2_Pct',
    'Bene_CC_PH_HF_NonIHD_V2_Pct', 'Bene_CC_PH_Hyperlipidemia_V2_Pct', 'Bene_CC_PH_Hypertension_V2_Pct',
    'Bene_CC_PH_IschemicHeart_V2_Pct', 'Bene_CC_PH_Osteoporosis_V2_Pct', 'Bene_CC_PH_Parkinson_V2_Pct',
    'Bene_CC_PH_Arthritis_V2_Pct', 'Bene_CC_PH_Stroke_TIA_V2_Pct', 'Bene_Avg_Risk_Scre', 'Year'
]

# drop duplicates
df_provider = df[provider_cols].drop_duplicates(subset=['Rndrng_NPI', 'Year']).reset_index(drop=True)
df_address = df[address_cols].drop_duplicates(subset=['Rndrng_NPI']).reset_index(drop=True)
df_ruca = df[ruca_cols].drop_duplicates(subset=['Rndrng_Prvdr_RUCA']).reset_index(drop=True)
df_state = df[state_cols].drop_duplicates(subset=['Rndrng_Prvdr_State_Abrvtn']).reset_index(drop=True)
df_classification = df[classification_cols].drop_duplicates(subset=['Rndrng_NPI', 'Year']).reset_index(drop=True)
df_service_totals = df[service_totals_cols].drop_duplicates(subset=['Rndrng_NPI', 'Year']).reset_index(drop=True)
df_drug_services = df[drug_services_cols].drop_duplicates(subset=['Rndrng_NPI', 'Year']).reset_index(drop=True)
df_medical_services = df[medical_services_cols].drop_duplicates(subset=['Rndrng_NPI', 'Year']).reset_index(drop=True)
df_bene_demo = df[bene_demo_cols].drop_duplicates(subset=['Rndrng_NPI', 'Year']).reset_index(drop=True)
df_bene_cc = df[bene_cc_cols].drop_duplicates(subset=['Rndrng_NPI', 'Year']).reset_index(drop=True)
```

```python
# Export each DataFrame to CSV
df_provider.to_csv('provider.csv', index=False)
df_address.to_csv('provider_address.csv', index=False)
df_ruca.to_csv('ruca.csv', index=False)
df_state.to_csv('state.csv', index=False)
df_classification.to_csv('provider_classification.csv', index=False)
df_service_totals.to_csv('service_totals.csv', index=False)
df_drug_services.to_csv('drug_services.csv', index=False)
df_medical_services.to_csv('medical_services.csv', index=False)
df_bene_demo.to_csv('beneficiary_demographics.csv', index=False)
df_bene_cc.to_csv('beneficiary_conditions.csv', index=False)

print("All normalized tables exported successfully!")
```
