```python
#Here we are creating a dataframe for the year 2023
#We can consider the dataframe (df_2023) as a table where all the data from Geography_and_Service_2023 has been loaded. 
file_name_Geography_2023 = 'MUP_PHY_R25_P05_V20_D23_Geo.csv'
df_2023 = pd.read_csv(file_name_Geography_2023, low_memory=False)

# Here we are adding column (year) to our 2023's dataset
df_2023['Year'] = 2023

#Here seeing the column names and first 3 rows.
df_2023.head(3) 
```

```python
import numpy as np
import pandas as pd

# Mapping
us_state_abbrev = {
    'Alabama': 'AL', 'Alaska': 'AK', 'Arizona': 'AZ', 'Arkansas': 'AR',
    'California': 'CA', 'Colorado': 'CO', 'Connecticut': 'CT', 'Delaware': 'DE',
    'Florida': 'FL', 'Georgia': 'GA', 'Hawaii': 'HI', 'Idaho': 'ID',
    'Illinois': 'IL', 'Indiana': 'IN', 'Iowa': 'IA', 'Kansas': 'KS',
    'Kentucky': 'KY', 'Louisiana': 'LA', 'Maine': 'ME', 'Maryland': 'MD',
    'Massachusetts': 'MA', 'Michigan': 'MI', 'Minnesota': 'MN', 'Mississippi': 'MS',
    'Missouri': 'MO', 'Montana': 'MT', 'Nebraska': 'NE', 'Nevada': 'NV',
    'New Hampshire': 'NH', 'New Jersey': 'NJ', 'New Mexico': 'NM', 'New York': 'NY',
    'North Carolina': 'NC', 'North Dakota': 'ND', 'Ohio': 'OH', 'Oklahoma': 'OK',
    'Oregon': 'OR', 'Pennsylvania': 'PA', 'Rhode Island': 'RI', 'South Carolina': 'SC',
    'South Dakota': 'SD', 'Tennessee': 'TN', 'Texas': 'TX', 'Utah': 'UT',
    'Vermont': 'VT', 'Virginia': 'VA', 'Washington': 'WA', 'West Virginia': 'WV',
    'Wisconsin': 'WI', 'Wyoming': 'WY'
}

valid_states = set(us_state_abbrev.keys())

# Keep only rows with valid U.S. states
df = df[df['Rndrng_Prvdr_Geo_Desc'].str.strip().str.title().isin(valid_states)].copy()

# Function to map state names to abbreviations
def extract_state_abbr(geo_desc):
    if pd.isna(geo_desc):
        return np.nan
    geo_desc_clean = geo_desc.strip().title()
    return us_state_abbrev.get(geo_desc_clean, np.nan)

# Apply function to create abbreviation column
df['Rndrng_Prvdr_State_Abrvtn'] = df['Rndrng_Prvdr_Geo_Desc'].apply(extract_state_abbr)
```

```python
#If a cell contains a missing value (like NaN or None), it returns True.
#If a cell contains a valid value, it returns False.
#To understand which column has null value or not we can write this funtion and print. 
#(We might need to understand this when we will decide to write query/ might need to handle null values).

import numpy as np

# replace empty strings with NaN
df.replace('', np.nan, inplace=True)

# check columns with any nulls
null_columns = df.columns[df.isnull().any()]
print("Columns with null values:", null_columns.tolist())

# count of nulls per column
print(df.isnull().sum()[df.isnull().sum() > 0])
```

```python
# Geography Table
geography_cols = [
   'Rndrng_Prvdr_State_Abrvtn','Rndrng_Prvdr_Geo_Cd', 'Rndrng_Prvdr_Geo_Lvl', 'Rndrng_Prvdr_Geo_Desc'
]

# HCPCS Table
hcpcs_cols = [
    'HCPCS_Cd', 'HCPCS_Desc', 'HCPCS_Drug_Ind', 'Place_Of_Srvc'
]

# Benefit Service and Amount Costs per state
benefit_service_amount_cols = [
    'Rndrng_Prvdr_State_Abrvtn', 'HCPCS_Cd', 'Year', 'Tot_Rndrng_Prvdrs',
    'Tot_Benes', 'Tot_Srvcs', 'Tot_Bene_Day_Srvcs', 'Avg_Sbmtd_Chrg',
    'Avg_Mdcr_Alowd_Amt', 'Avg_Mdcr_Pymt_Amt', 'Avg_Mdcr_Stdzd_Amt'
]

# drop duplicates
df_geography = df[geography_cols].drop_duplicates(subset=['Rndrng_Prvdr_State_Abrvtn']).reset_index(drop=True)
df_hcpcs = df[hcpcs_cols].drop_duplicates(subset=['HCPCS_Cd']).reset_index(drop=True)
df_benefit_service_amount = df[benefit_service_amount_cols].drop_duplicates(subset=['Rndrng_Prvdr_State_Abrvtn', 'Year']).reset_index(drop=True)
```

```python
# Export each DataFrame to CSV
df_geography.to_csv('geography.csv', index=False)
df_hcpcs.to_csv('hcpcs_codes1.csv', index=False)
df_benefit_service_amount.to_csv('state_providers_benefit_service_amounts.csv', index=False)

print("All normalized tables exported successfully!")
```
